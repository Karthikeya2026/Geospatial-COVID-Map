```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import geodatasets

# Load world countries map using geodatasets
world = gpd.read_file(geodatasets.get_path('naturalearth_lowres'))

# Sample COVID-19 data (simplified manually for demo purposes)
covid_data = pd.DataFrame({
    'country': ['China', 'United States of America', 'Brazil', 'India', 'Russia'],
    'cases': [90000, 500000, 400000, 300000, 200000]
})

# Merge world map with COVID data
merged = world.merge(covid_data, how='left', left_on='name', right_on='country')

# Plot
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged.plot(column='cases', cmap='Reds', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Add title
plt.title('COVID-19 Cases by Country (Demo Map)', fontsize=15)
plt.axis('off')
plt.show()
```
